Using $k = \sqrt{(kc)^2 + \pi^2c^2\left[\left(\frac{m}{a}\right)^2 + \left(\frac{n}{b}\right)^2\right]}$

```python
import numpy as np
import matplotlib.pyplot as plt
#
def omega(k, a, b, m, n):
    return np.sqrt((k*c)**2 + np.pi**2*c**2*((m/a)**2 + (n/b)**2))
#
k = np.linspace(0, 1000, 1000)
c=3.00e8
a = 0.0228 # height 
b = 0.0101 # width
#
plt.figure(figsize=(12,10))
plt.plot(k,c*k,label=r"$\omega=ck$")
plt.plot(k,omega(k,a,b,1,0),label="TE10")
plt.plot(k,omega(k,a,b,2,0),label="TE20")
plt.plot(k,omega(k,a,b,3,0),label="TE30")
plt.plot(k,omega(k,a,b,0,1),label="TE01")
plt.plot(k,omega(k,a,b,0,2),label="TE02")
plt.plot(k,omega(k,a,b,1,1),label="TE11")
plt.plot(k,2*np.pi*17e9*np.ones(1000),label=r"$\omega$ = (2$\pi)$17e9 rad/s")
plt.title('Dispersion Relation')
plt.xlabel(r'$k$ (1/m)')
plt.ylabel(r'$\omega$ (rad/s)')
plt.legend()
```

---
The phase velocity is given by $v = \omega/k =c/\sqrt{1-(\omega_{mn}/\omega)^2}$. For the frequencies that are allowed to propagate ($\omega > \omega_{mn}$) $v$ is larger than $c$ but slows down and approaches it in the limit $\omega >> \omega_{mn}$.

The group velocity is given by $v_g = \frac{1}{dk/d\omega} = c\sqrt{1 - (\omega_{mn}/\omega)^2} < c$. $v_g$ increases and approaches $c$ in the limit $\omega >> \omega_{mn}$.

Only the $\text{TE}_{10}$, $\text{TE}_{20}$, $\text{TE}_{01}$ and $\text{TE}_{11}$ modes are allowed. This can be seen on the plot if we look at the lines that intersect the cutoff frequency $\omega_{mn} = (2\pi)17 \times 10^9$rad/s. The $\text{TE}_{30}$ and $\text{TE}_{02}$ modes begin at frequencies higher than the cutoff frequency at $k$ = 0 and never intersect the line.
